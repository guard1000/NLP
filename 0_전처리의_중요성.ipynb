[1_MLP.ipynb], [2_keras_reuters_application.ipynb] 에서 사용한 '전처리가 잘 되어있던' .json파일의 데이터 구조는
data_json['X'] : 리스트로 [[56681개] [56681개] [56681개의 0~238인 수 ] ... 8419개]
-- 이건 56681개의 vocab들중 0번이 몇번 나타났고~ 의 정보를 담고 있음
-- 즉, BoW 되어있는 것임
-- 이건 92% 까지도 나옴


아래 코드는 1의 데이터를 2의 처리방식으로 했을때 결과.
-- 역시 79점따리 나옴
-- 전처리를 얼마나 잘하는지가 중요하다

```
###################################################
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

train_data = pd.read_csv("/content/drive/My Drive/data/train_data.csv")
test_data = pd.read_csv("/content/drive/My Drive/data/test_data.csv")
```

```
X_train = []
X_test = []

for i in range(len(train_data)):
  X_train.append(train_data.iloc[i].Query.split())
for i in range(len(test_data)):
  X_test.append(test_data.iloc[i].Query.split())

print(len(X_train), len(X_test))
print(type(X_train), type(X_test))
print(X_train[0])
print(X_test[0])
```

```
X_total = X_train + X_test

tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_total)
tokenizer.word_index
```

```
total_cnt = len(tokenizer.word_index) # 단어의 수
total_cnt
```

```
# BoW용 벡터 만들기
new_X_train = np.zeros((10000,69180))
new_X_test = np.zeros((1000,69180))
```

```
new_X_test.shape
```

```

for i in range(len(X_train)):
  for j in X_train[i]:
    try:
      new_X_train[i][tokenizer.word_index[j]-1] += 1
    except:
      pass

for i in range(len(X_test)):
  for j in X_test[i]:
    try:
      new_X_test[i][tokenizer.word_index[j]-1] += 1
    except:
      pass

print(new_X_train[0])
```

```
cat_dic = {'경제': 0, '생활문화': 1, 'IT과학': 2, '정치': 3, '사회': 4, '오피니언': 5, '세계':6}
Y_train = []
Y_test = []


for i in range(len(train_data)):
  Y_train.append(cat_dic[train_data.iloc[i].GT])
for i in range(len(test_data)):
  Y_test.append(cat_dic[test_data.iloc[i].GT])

print(len(Y_train), len(Y_test))
print(type(Y_train), type(Y_test))
print(Y_train[:10])
print(Y_test[:10])
```

```
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
from sklearn import model_selection, metrics
max_words = 69180 # 입력 단어 수: word-dic.json 파일 참고
nb_classes = 7 # 7개 카테고리 경제, 생활문화, IT과학, 정치, 사회, 오피니언, 세계
batch_size = 64
nb_epoch = 10

# build MLP model
def build_model():
    model = Sequential()
    model.add(Dense(512, input_shape=(max_words,))) # dense : 전결합 신경망, 출력 차원 512, 입력차원 56681
    model.add(Activation('relu')) # 활성화 함수
    model.add(Dropout(0.5)) #드롭아웃
    model.add(Dense(nb_classes)) # 출력차원 : 7
    model.add(Activation('softmax'))
    model.compile(loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy'])
    return model
```

```
model = KerasClassifier(
    build_fn=build_model,
    nb_epoch=nb_epoch,
    batch_size=batch_size)
model.fit(np.array(new_X_train),np.array(Y_train))
```

```
y = model.predict(np.array(new_X_test))
ac_score = metrics.accuracy_score(Y_test, y)
cl_report = metrics.classification_report(Y_test, y)
print("정답률 =", ac_score)
print("리포트 =\n", cl_report)
```

```
# 모델 만들기
# 필요 라이브러리들 소환
from tensorflow.keras import models
from tensorflow.keras import layers
from sklearn import metrics

model2 = models.Sequential()
model2.add(layers.Dense(64, activation='relu', input_shape=(max_words, )))
model2.add(layers.Dense(64, activation='relu'))
model2.add(layers.Dense(7, activation='softmax'))

model2.summary()
```

```
model2.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'],
)
```

```
x_val = new_X_train[:1000]
partial_x_train = new_X_train[1000:]
y_val = Y_train[:1000]
partial_y_train = Y_train[1000:]

history = model2.fit(
    partial_x_train, 
    partial_y_train,
    epochs=10,
    batch_size=512,
    validation_data=(x_val, y_val)
)
```

```
predictions = model2.predict(new_X_test)
answer = [np.argmax(i) for i in predictions]

ac_score = metrics.accuracy_score(Y_test, answer)
cl_report = metrics.classification_report(Y_test, answer)
print("정답률 =", ac_score)
print("리포트 =\n", cl_report)
```
